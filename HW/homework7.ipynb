## Problem: Rapidity and the Lorentz-Invariant Phase Space

The quantity  

$ \displaystyle \frac{d^3p}{E} $

is a Lorentz-invariant element of phase space. In high-energy collisions, it is often convenient to express this measure in terms of the particle’s **rapidity** and **transverse momentum**.

The **rapidity** $y$ is defined by  

$ y = \frac{1}{2}\ln\!\left(\frac{E + p_z}{E - p_z}\right), $

where $E$ and $p_z$ are the energy and the longitudinal (beam-axis) component of the particle’s momentum, respectively.

The **transverse momentum** $p_T$ is defined as  

$ p_T = \sqrt{p_x^2 + p_y^2}, $

i.e., the magnitude of the momentum in the plane perpendicular to the beam direction.

## Solutions: Rapidity and the Lorentz-Invariant Phase Space

---

### (a) Expressing $E$ and $p_z$ in terms of $y$, $p_T$, and $m$

Starting from the definition of rapidity:

$ y = \frac{1}{2}\ln\!\left(\frac{E + p_z}{E - p_z}\right) $

we can write

$ e^{2y} = \frac{E + p_z}{E - p_z}. $

Solving for $E \pm p_z$ gives:

$ E + p_z = (E - p_z)e^{2y}. $

Rearranging terms:

$ E(1 - e^{2y}) = -p_z(1 + e^{2y}). $

We can also note that

$ (E + p_z)(E - p_z) = E^2 - p_z^2 = p_T^2 + m^2. $

Then:

$ (E + p_z)(E - p_z) = p_T^2 + m^2. $

Using $E + p_z = \sqrt{p_T^2 + m^2}\, e^{y}$ and $E - p_z = \sqrt{p_T^2 + m^2}\, e^{-y}$,  
we can solve for $E$ and $p_z$:

**Answer:**

$ E = \sqrt{p_T^2 + m^2}\,\cosh y, \quad p_z = \sqrt{p_T^2 + m^2}\,\sinh y. $

---

### (b) Expressing $d^3p/E$ in terms of $y$, $\phi$, and $p_T$

In cylindrical coordinates, the momentum-space volume element is

$ d^3p = p_T\,dp_T\,d\phi\,dp_z. $

At fixed $p_T$, we have $p_z = \sqrt{p_T^2 + m^2}\,\sinh y$, so

$ dp_z = \sqrt{p_T^2 + m^2}\,\cosh y\,dy. $

Substituting this into $d^3p$ gives

$ d^3p = p_T\,dp_T\,d\phi\,(\sqrt{p_T^2 + m^2}\,\cosh y\,dy). $

Since $E = \sqrt{p_T^2 + m^2}\,\cosh y$, dividing by $E$ simplifies the expression:

$ \frac{d^3p}{E} = p_T\,dp_T\,d\phi\,dy. $

We can also write $p_T\,dp_T = \tfrac{1}{2}\,d(p_T^2)$, so equivalently:

**Answer:**

$ \frac{d^3p}{E} = \tfrac{1}{2}\,d(p_T^2)\,d\phi\,dy. $

This shows explicitly that the phase-space element $d^3p/E$ can be written in terms of rapidity, azimuthal angle, and transverse momentum.

---

### (c) The pseudorapidity $\eta$ in the massless limit

In the **massless limit** ($m \to 0$), $E = |\vec{p}| = p$, and we can define the **pseudorapidity** $\eta$ as:

$ \eta = \frac{1}{2}\ln\!\left(\frac{p + p_z}{p - p_z}\right). $

Using the relation between the polar angle $\theta$ and the momentum components:

$ p_z = p\cos\theta, \quad p_T = p\sin\theta, $

we find:

$ \eta = \frac{1}{2}\ln\!\left(\frac{1 + \cos\theta}{1 - \cos\theta}\right)
= \frac{1}{2}\ln\!\left(\frac{\cos^2(\theta/2)}{\sin^2(\theta/2)}\right)
= -\ln\!\big[\tan(\tfrac{\theta}{2})\big]. $

**Answer:**

$ \eta = -\ln\!\big[\tan(\tfrac{\theta}{2})\big]. $

This expression is widely used in collider physics, as $\eta$ depends only on the polar angle and is invariant under boosts along the beam axis.


## Problem: Collider Coordinate System

The figure below illustrates the **Collider Detector Coordinate System**. The incoming beam axis is typically labeled as the **z-axis**, which defines the direction of the particle beams in a collider. The **x-axis** and **y-axis** are defined to be perpendicular to the z-axis, forming what is known as the **transverse plane**.

The right-pointing red arrow represents the **final-state particle**, which is produced from the collision. The projection of the final-state particle’s momentum onto the transverse plane is called the **transverse momentum**, denoted as $p_T$.

The angle between the final-state particle’s trajectory and the beam line (z-axis) is called the **polar angle**, denoted as $\theta$. The angle between the transverse momentum $p_T$ and the x-axis is called the **azimuthal angle**, denoted as $\phi$.

As shown in the previous problem, the polar angle $\theta$ has a relationship with the **pseudorapidity**, $\eta$. In collider physics, $\eta$ is often used instead of $\theta$ because pseudorapidity is **Lorentz invariant**, making it a more convenient quantity for describing particle kinematics in high-energy collisions.

The Deep Inelastic Scattering cross sections can be used to determine the distribution of partons inside the proton.  The parton distribution functions measured this way can be used in other hadron scattering processes.  For example, consider the process

$$
p p \rightarrow \mu^+ \mu^- X
$$

The cross section for this process can be calculated by convoluting the cross section

$$
\sigma(q\overline q \rightarrow \mu^+\mu^-)
$$

with the appropriate probability distribution functions for finding quark $q$ in one proton and and anti-quark $\overline q$ in the other proton.  A good picture of this process (taken from Wikpedia) is shown below.  The $q$ amd $\overline q$ annihilate into a virtual photon that then turns into the lepton pair.

The cross section for the reaction $pp\rightarrow \mu^+\mu-+X$ is given by

$$                                                         
\frac{d\sigma(p p\rightarrow \mu^+ \mu^- X)}{d\hat s} = 
\sum_{ij} f_i^{(q)} (x_1) f_j^{(\overline q)} (x_2) \frac{d\sigma(q_i\overline q_j\rightarrow \mu^+ \mu^- X)}{d\hat s}
$$

where $\hat s$ is the square of the $\mu^+\mu^-$ invariant mass, $f_i^{(q)}(x_1)$ is the parton distribution function for quarks of species $i$ to carry a fraction $x_1$ of the proton's momentum and  $f_y^{(\overline q)}(x_2)$ is the parton distribution function for anti-quarks of species $i$ to carry a fraction $x_2$ of the proton's momentum. (this is the same $f_i(x)$ as in the previous problem)  Note that for the process above $i=j$ since the quark and anti-quark must be of the same species.
