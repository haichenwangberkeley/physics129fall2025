# The BaBar Experiment: Calculating Center-of-Mass Energy

### Problem Description

At the **BaBar experiment**, an electron beam ($e^-$) with an energy of **9.0 GeV** was used to collide with a positron beam ($e^+$). The goal was to create a specific particle, the **$\Upsilon(4S)$ meson**, which has a known rest mass of **10.58 GeV**. The challenge is to figure out what energy the positron beam must have to achieve this specific center-of-mass energy.



### Your Task

Given a **9.0 GeV electron beam** ($E_1$) and a target $\Upsilon(4S)$ mass of **10.58 GeV** ($E_{CM}$), what should be the energy ($E_2$) of the positron beam in a head-on collision?

## ‚úÖ Physics Mini-Examples

**Electron / Positron mass**  
Inline: The electron mass is $m_e \approx 0.511~\text{MeV}/c^2$.  
Display:
$$
m_{e^+} = m_e \approx 0.511~\text{MeV}/c^2
$$

**Energy‚Äìmomentum relation (natural units)**
$$
E^2 = p^2 + m^2
$$

**Center-of-mass energy (symmetric high-energy collider)**
$$
E_{\text{CM}} \approx 2E \quad (\text{for } E_1 = E_2 = E \gg m)
$$


**Breit‚ÄìWigner function)**

The Breit‚ÄìWigner distribution, used to describe resonances, is

$$
\sigma(E) = \frac{\Gamma^2 / 4}{(E - E_0)^2 + \Gamma^2 / 4}
$$

where  
- $E$ is the energy,  
- $E_0$ is the resonance energy (peak position),  
- $\Gamma$ is the decay width of the resonance.
